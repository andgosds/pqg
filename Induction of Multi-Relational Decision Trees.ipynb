## Learning Induction Trees from Multi-relational DataBases in Neo4J

We present the implementation of our own multi-relational decision tree learning algorithm. This implementation has been developed in Python and is able to automatically induce trees from structures (subgraphs) stored in a Neo4j database. The algorithm use Property Query Graphs: a multi-relational pattern that acts as an attribute in the decision tree construction process. Such PQGs are built dynamically during the tree construction process: the algorithm performs pattern mining all while learning a classification tree for the structures stored in Neo4j.

Loading libraries and configuring matplotlib to plot inline...

Building model, notice that the target field should be present in every node in the training set...

```python
modelo = id3(7474,"neo4j","pytpytpyt","gender") #id3(port,neo4j_username,neo4j_password,target_field)
```

Creating a starting PQG composed by one empty node (with id "0")...

```python
q = PropertyQueryGraph()
q.addNode("0",True,{},True)
```

Learning the decision tree from a training set of nodes (represented by their neo4j id)...

```python
tree = modelo.execute(q , [[1],[3],[4],[6],[8],[9],[11],[13]])
```
